```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
import pandas as pd
```

```python
url='https://www.nike.com/il/w/womens-sale-tops-t-shirts-3yaepz5e1x6z9om13'
browser=webdriver.Chrome()
browser.get(url)
print(browser.title)
browser.find_element(By.XPATH,'/html/body/div[3]/div/div[1]/div/div[2]/div/div[2]/div[2]/button').click()

body_elem=browser.find_element(By.TAG_NAME,'body')
scrool=100
while scrool:
    body_elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
    scrool-=1


products=browser.find_elements(By.CLASS_NAME,'product-card__title')
atags=browser.find_elements(By.CLASS_NAME,'product-card__link-overlay')


#________________
products_name=[product.text for product in products]
links=[link.get_attribute('href') for link in atags]

time.sleep(3)
browser.close()
```

```python
pd.DataFrame({'Product':products_name,'Links':links})
```
